Simulate a discrete probability distribution P = [0.2, 0.5, 0.3] and Q = [1/3, 1/3, 1/3]

Compute Cross Entropy and KL-Divergence

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the distributions
P = [0.2, 0.5, 0.3]
Q = [1/3, 1/3, 1/3]
outcomes = [0, 1, 2]
n_samples = 10000

# Set random seed for reproducibility
np.random.seed(42)

# Generate samples
samples_P = np.random.choice(outcomes, size=n_samples, p=P)
samples_Q = np.random.choice(outcomes, size=n_samples, p=Q)

# Calculate empirical probabilities
hist_P, _ = np.histogram(samples_P, bins=[-0.5, 0.5, 1.5, 2.5], density=True)
hist_Q, _ = np.histogram(samples_Q, bins=[-0.5, 0.5, 1.5, 2.5], density=True)

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot for P
ax1.bar(outcomes, P, width=0.45, alpha=0.7, label='Theoretical P', color='blue')
ax1.bar(np.array(outcomes) + 0.45, hist_P, width=0.45, alpha=0.7, label='Empirical P', color='lightblue')
ax1.set_title('Distribution P')
ax1.set_xlabel('Outcome')
ax1.set_ylabel('Probability')
ax1.set_xticks(outcomes)
ax1.legend()

# Plot for Q
ax2.bar(outcomes, Q, width=0.45, alpha=0.7, label='Theoretical Q', color='red')
ax2.bar(np.array(outcomes) + 0.45, hist_Q, width=0.45, alpha=0.7, label='Empirical Q', color='pink')
ax2.set_title('Distribution Q')
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Probability')
ax2.set_xticks(outcomes)
ax2.legend()

plt.tight_layout()
plt.show()

# Print sample counts
print("Sample counts for P:", np.bincount(samples_P))
print("Sample counts for Q:", np.bincount(samples_Q))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

#LOADING MNIST DATASET


```python
#pre-processor
transform =  transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5),(0.5))
])
'''Key Points to Remember
transforms.Compose chains multiple image transformations together.

ToTensor converts images to PyTorch tensors and scales pixel values to [0, Normalize((0.5), (0.5)) shifts and scales tensor values to the range [-1, 1].

This normalization is often used for training neural networks, as it helps with model convergence and stability.'''
#load the data
train_dataset =  datasets.MNIST('data' , train = True, download = True, transform = transform)
test_dataset =  datasets.MNIST('data', train = False , download = True , transform = transform)

'''data': The directory where the dataset will be stored (downloaded if not present).
train=True: Loads the training set (60,000 images).

train=False: Loads the test set (10,000 images).
download=True: Downloads the dataset if it isn't already present in the specified directory.'''

#make data loaders

train_loader =  torch.utils.data.DataLoader(train_dataset, batch_size =32 , shuffle = True)
test_loader  =  torch.utils.data.DataLoader(test_dataset, batch_size  = 32, shuffle = False)
'''torch.utils.data.DataLoader: A PyTorch utility that loads data in batches and optionally shuffles it.'''
```

```python
# If you want a quick visualization of a batch:

images, labels = next(iter(train_loader))
print(images.shape)  # torch.Size([32, 1, 28, 28])
print(labels.shape)  # torch.Size([32])
```

```python
#Build a model
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        #in_channel , out_channel (activation map), filter_size , stride
        # iNPUT (1, 28, 28)
        self.conv1 = nn.Conv2d(1, 32, 3, 1) # (N-F)/S + 1 = (28 - 3)/ 1 +1 = 26 (32, 26, 26)
        self.conv2 = nn.Conv2d(32, 64, 3, 1) # (26-3)/1 +1 = 24/2  (MaxPool); (64, 12, 12) = 9216
        self.fc1 = nn.Linear(9216,1024)
        self.fc = nn.Linear(1024, 10)
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = self.fc(x)
        output = F.log_softmax(x, dim=1)
        return output

def initialize_model():
    model = CNNModel()
    optimizer = optim.SGD(model.parameters(), lr=0.01)
    loss_fun = nn.CrossEntropyLoss()
    return model, loss_fun, optimizer
```

```python
mnist_model, loss_fn, optimizer = initialize_model()
```

```python
mnist_model
```

```python
def train_model(model, train_loader, test_loader, loss_fn, optimizer, epochs=10):
    train_acc, test_acc = [], []
    losses = torch.zeros(epochs)

    for epoch in range(epochs):
        model.train()
        epoch_losses, epoch_accuracies = [], []

        for X_batch, y_batch in train_loader:
            y_pred = model(X_batch)#Forward pass
            loss = loss_fn(y_pred, y_batch)

            optimizer.zero_grad()#Clear previous gradients
            loss.backward() #Backward pass (compute gradient
            optimizer.step()#update the model paramters

            epoch_losses.append(loss.item())
            acc = (torch.argmax(y_pred, axis=1) == y_batch).float().mean() * 100
            epoch_accuracies.append(acc.item())

        model.eval()
        X_test, y_test = next(iter(test_loader))
        with torch.no_grad():
            y_test_pred = model(X_test)
        test_accuracy = (torch.argmax(y_test_pred, axis=1) == y_test).float().mean() * 100

        train_acc.append(np.mean(epoch_accuracies))
        test_acc.append(test_accuracy.item())
        losses[epoch] = np.mean(epoch_losses)

    return train_acc, test_acc, losses, model
```

```python
# prompt: continue the code to run model
import numpy as np

train_acc, test_acc, losses, trained_model = train_model(mnist_model, train_loader, test_loader, loss_fn, optimizer, epochs=5)

print(f"PyTorch Test Accuracy: {test_acc[-1]:.4f}")

```

```python
print(f'train accuracy : {train_acc}')
print(f'test accuracy : {test_acc}')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# scalling
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# TensorFlow + Keras implementation
# Define the CNN model
keras_model = Sequential([
    tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the model
keras_model.compile(optimizer='sgd',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

# Prepare the data for Keras (add channel dimension)
x_train_keras = x_train.reshape(-1, 28, 28, 1)
x_test_keras = x_test.reshape(-1, 28, 28, 1)

# Train the Keras model
keras_model.fit(x_train_keras, y_train, epochs=5, validation_data=(x_test_keras, y_test))

# Evaluate the Keras model
loss, accuracy = keras_model.evaluate(x_test_keras, y_test, verbose=0)
print(f"Keras Test Accuracy: {accuracy:.4f}")

# Print model summaries
print("\nPyTorch Model Summary:")
print(mnist_model)

print("\nKeras Model Summary:")
keras_model.summary()
```
